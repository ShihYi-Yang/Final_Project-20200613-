```python
# 新增
def add_player(players_data):
    player_name = input("輸入球員姓名: ")
    if player_name in players_data:   # 檢查球員是否在 dict 內
        print(f"{player_name} 已存在，不必新增.")
    else:
        comment = ["所屬球隊", "生涯場均得分", "生涯場均籃板", "生涯場均助攻", "總冠軍數", "入選明星賽次數"]
        information = []
        
        for i in range(len(comment)):
            information.append(input(f"{comment[i]}:"))
            
        for i in range(len(information)):   # 把 string 變成 int or float, 後面才可比較
            if i >=1 and i <= 3:
                information[i] = float(information[i])
            elif i > 3:
                information[i] = int(information[i])
            
        players_data[player_name] = information
        
        print(f"{player_name} 已加入資料中.")
        
        
# 刪除
def delete_player(NBA):
    player_name = input("輸入球員姓名: ")
    if player_name in players_data:
        choice = input("如確定要刪除，請輸入yes，不刪除請輸入no : ")
        if choice == "yes":
            del players_data[player_name]
            print(f"已刪除 {player_name} 之資料.")
        else:
            print(f"並未刪除 {player_name} 之資料.")
    else:
        print(f"{player_name} 不存在資料中.")
        
        
# 查詢
def search_player(players_data):
    player_name = input("輸入球員姓名: ")
    if player_name in players_data:
        comment = ["所屬球隊", "生涯場均得分", "生涯場均籃板", "生涯場均助攻", "總冠軍數", "入選明星賽次數"]
        
        for i in range(len(comment)):
            print(f"{comment[i]}: {players_data[player_name][i]}")
    else:
        print(f"{player_name} 不存在資料中.")
        
        
# 修改
def modify_player_data(players_data):
    player_name = input("輸入球員姓名: ")
    if player_name in players_data:
        comment = ["所屬球隊", "生涯場均得分", "生涯場均籃板", "生涯場均助攻", "總冠軍數", "入選明星賽次數"]
        information = []
        
        for i in range(len(comment)):
            information.append(input(f"{comment[i]}:"))
            
        for i in range(len(information)):   # 把 string 變成 int or float, 後面才可比較
            if i >=1 and i <= 3:
                information[i] = float(information[i])
            elif i > 3:
                information[i] = int(information[i])
        
        players_data[player_name] = information
        
        print(f"{player_name} 之資料修改成功.")
    else:
        print(f"{player_name} 不存在資料中.")
        
        
# 列印出各項數據前十名
def top_ten_player(players_data):
    index = int(input("輸入你要查詢的選項，查詢生涯場均得分輸入1、生涯場均籃板輸入2、生涯場均助攻輸入3、總冠軍數輸入4、入選明星賽次數輸入5"))
    
    reverse_Dict = dict()
    for player in players_data:
        temp = players_data[player][index]   # 5個數據中之其中一個, ex：Lebron James 的場均得分
        if temp not in reverse_Dict:
            reverse_Dict[temp] = [player]   # {27.2 : [Lebron James]}
        else:
            reverse_Dict[temp].append(player)   # 以防球員資料一樣
            
    target_list = [key for key in reverse_Dict.keys()]
    target_list.sort(reverse=True)   # 數值高的排前面
    
    for i in range(10):
        if target_list[i] == 0:   # 有些項目排不到前10名
            return None
        print(f"{i+1}. { ', '.join(reverse_Dict[target_list[i]]) }: {target_list[i]}")
        

        
# 比較兩個球員之數據        
def compare(players_data):
    player1 = input("輸入第一位球員: ") 
    if player1 not in players_data:
        print(f"{player1} 不存在資料中.")
        return None
    
    player2 = input("輸入第二位球員: ")
    if player2 not in players_data:
        print(f"{player2} 不存在資料中.")
        return None
    
    player1_list, player2_list = players_data[player1][1:], players_data[player2][1:]   # 去掉所屬球隊
    comment = ["生涯場均得分", "生涯場均籃板", "生涯場均助攻", "總冠軍數", "入選明星賽次數"]
    
    print("")
    for i in range(len(comment)):
        print(f"{comment[i]}: ", end="")
        if player1_list[i] > player2_list[i]:
            print(f"{player1} 勝.")
        elif player1_list[i] < player2_list[i]:
            print(f"{player2} 勝.")
        else:
            print("平手.")
```
